```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def run(x, trajectory):
    trajectory.append(x)
    if x == 1:
        return np.array(trajectory)
    elif x % 2 == 0:
        run(x/2, trajectory)
    elif x % 2 == 1:
        run(3 * x + 1, trajectory)
    return np.array(trajectory)
```

```python
for i in range(1000):
    plt.plot(run(i+1, []))
plt.show()
```
